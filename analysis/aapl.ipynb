```python
import polars as pl
import seaborn as sns
import plotly.graph_objects as go
```

```python
btc = pl.read_csv("../data/BTC_USD_all_time.csv")
btc.head()
```

```python
fig = go.Figure(
    data=[
        go.Candlestick(
            x=btc["timestamp"],
            open=btc["open"],
            high=btc["high"],
            low=btc["low"],
            close=btc["close"],
        )
    ]
)

# Update layout
fig.update_layout(
    title="btc Candlestick Chart",
    xaxis_title="Date",
    yaxis_title="Price",
    xaxis_rangeslider_visible=False,
)

# Show the figure
fig.show()
```
